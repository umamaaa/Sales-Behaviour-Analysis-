```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\PMLS\Documents\Project\Customer Purchasing Behaviors.csv", on_bad_lines='skip')
```

```python
#CORRELATION B/W COLUMNS
n_col = ['age', 'annual_income', 'loyalty_score', 'purchase_amount', 'purchase_frequency']

correlation_matrix = df[n_col].corr().round(2)

plt.figure(figsize=(8, 8))  # Set the figure size
sns.heatmap(correlation_matrix, annot=True, cmap="Blues", square=True, cbar=True, linewidths=0.5)

plt.show()
```

```python
sns.scatterplot(data=df, x='annual_income', y='purchase_amount', hue='region', size='loyalty_score', sizes=(20, 200))
plt.title('Annual Income vs Purchase Amount')
plt.show()
```

```python
#Box plot
sns.boxplot(data=df, x='region', y='purchase_amount')
plt.title('Average Purchase Amount by Region')
plt.show()
```

```python

# Bar plot
sns.barplot(data=df, x='loyalty_score', y='age')

plt.title('Stripe Plot of loyalty_score vs age')
plt.xlabel('loyalty_score')
plt.ylabel('age')

plt.xticks(rotation=45)

# Show plot
plt.show()
```
